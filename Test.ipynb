```python
import pandas as pd
import altair as alt
```

```python
vol_df = pd.read_csv('data/Volcanoes_cleaned.csv')
```

```python
vol_df_sum = vol_df.groupby('Country').agg('count')
```

```python
vol_df_sum = vol_df_sum.reset_index()
vol_df_sum.head(5)
```

```python
alt.Chart(vol_df_sum).mark_bar().encode(
        x=alt.X("Country:N").axis(labelAngle=0),
        y="Name:Q",
        shape='Origin'
    )
```
